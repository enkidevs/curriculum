```python
import matplotlib.pyplot as plt
```

```python
my_list = [4, 5, 5, 6, 7, 8, 9, 10, 14, 14, 16, 19]
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.hist(my_list, bins=10, range=(0, 20))
fig, ax = plt.subplots(figsize=(4,3))
ax.hist(my_list, bins=4, range=(0, 20))
plt.style.use('ggplot')
plt.show()
```
